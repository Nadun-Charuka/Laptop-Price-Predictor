```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("laptop_price.csv",encoding='latin1')
data.head(10)
```

```python
data["Ram"] = data["Ram"].str.replace('GB','').astype('int32')
data["Weight"] = data["Weight"].str.replace('kg','').astype('float32')
```

```python
print(data.dtypes)
```

```python
data.corr(numeric_only=True)['Price_euros']
```

```python
data['Company'].value_counts()
```

```python
def add_company(input_val):
    # Group the brands you want to categorize as "Other"
    others = ["Samsung", "Razer", "Mediacom", "Microsoft", "Xiaomi", "Vero", "Chuwi", "Google", "Fujitsu", "LG", "Huawei"]
    
    if input_val in others:
        return "Other"
    else:
        return input_val

data["Company"] = data["Company"].apply(add_company)
```

```python
data['Company'].value_counts()
```

```python
data["Product"].value_counts()
```

```python
len(data["Product"].value_counts())
```

```python
data["TypeName"].value_counts()
```

```python
data['Touchscreen'] = data["ScreenResolution"].apply(lambda x:1 if 'Touchscreen' in x else 0)
data['Ips'] = data["ScreenResolution"].apply(lambda x:1 if 'IPS' in x else 0)
```

```python
data["Cpu"]
```

```python
data['cpu_name'] = data['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
def set_processor(name):
    intelNames = ["Intel Core i3","Intel Core i5","Intel Core i7"]
    if name in intelNames:
        return name
    elif name.split()[0] == "AMD":
        return "AMD"
    else:
        return "Other"

data['cpu_name'] = data['cpu_name'].apply(set_processor)
```

```python
data['cpu_name'].value_counts()
```

```python
data['Gpu'].value_counts()
```

```python
data['gpu_name'] = data['Gpu'].apply(lambda x:''.join(x.split()[0:1]))
```

```python
data['gpu_name'].value_counts()
```

```python
data.shape
```

```python
#remove the ARM gpu row 
data = data[data['gpu_name'] != "ARM"]
```

```python
data.shape
```

```python
data.head()
```

```python
data['OpSys'].value_counts()
```

```python
def set_os(name):
    windowsNames = ["Windows 10","Windows 8","Windows 7"]
    macNames = ["Mac OS X", "macOS"]
    if name in windowsNames:
        return "Windows"
    elif name in macNames:
        return "Mac"
    elif name == "Linux":
        return "Linux"
    else:
        return "Other"

data['OpSys'] = data['OpSys'].apply(set_os)
```

```python
data['OpSys'].value_counts()
```

```python
data.head()
```

```python
data = data.drop(columns = ['laptop_ID','Product','Inches','ScreenResolution','Cpu','Gpu'])
```

```python
data.head()
```

```python
data = pd.get_dummies(data)
```

```python
data.head()
```

```python
data.shape
```

#begin the model traning

```python
x = data.drop('Price_euros',axis=1)
y =data['Price_euros']
```

```python
!pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
x_train.shape ,x_test.shape
```

```python
def model_acc(model):
    model.fit(x_train,y_train)
    acc = model.score(x_test,y_test)
    print(str(model) + ' --> ' + str(acc))
```

start training

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)

from sklearn.linear_model import Lasso
lasso = Lasso()
model_acc(lasso)

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_acc(dt)

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_acc(rf)
```

```python
#we will continue with only randomForestRegressor
from sklearn.model_selection import GridSearchCV
parameters = {'n_estimators': [10,50,100], 'criterion':['squared_error','absolute_error','poisson']}

grid_obj = GridSearchCV(estimator = rf, param_grid=parameters)
grid_fit = grid_obj.fit(x_train,y_train)
best_model = grid_fit.best_estimator_
best_model
```

```python
best_model.score(x_test,y_test)
```

```python
#save the model
import pickle
with open('predictor.pickle','wb') as file:
    pickle.dump(best_model,file)
```
